```python
import requests
import pandas as pd
import re
from selenium import webdriver
import t
```

```python
r = requests.get('https://tosdr.org/api/1/all.json')
json = r.json()
```

```python
regex = re.compile(r"tosdr\/review\/[\w\-\.]+")
tos = []
for key in json.keys():
    if re.match(regex, key):
        try:
            if len(json[key]["documents"]) > 0:
                tos.append(json[key]["documents"])
        except:
            pass
tos
```

```python
privacy = []
terms = []
for i in tos:
    for j in i:
        if "privacy" in j["name"].lower():
                 privacy.append(j["url"])
        else:
                 terms.append(j["url"])
```
